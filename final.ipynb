## Importing Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

###  Reading data

```python
dailyActivity_merged = pd.read_csv(r'..\Fitabase Data 4.12.16-5.12.16\dailyActivity_merged.csv')
```

### exploring data

```python
dailyActivity_merged.head()
```

```python
dailyActivity_merged.info()
```

#### changing datatype of ActivityDate to datetype

```python
dailyActivity_merged['ActivityDate'] = pd.to_datetime(dailyActivity_merged['ActivityDate'],format="%m/%d/%Y")
```

```python
dailyActivity_merged.info()
```

```python
dailyActivity_merged.head()
```

#### Extract day of the week from ActivityDate

```python
dailyActivity_merged['Day'] = dailyActivity_merged['ActivityDate'].dt.day_name()
```

```python
dailyActivity_merged.head()
```

### Analyzing data

```python
plt.figure(figsize=(20,15))
heatmap = sns.heatmap(dailyActivity_merged.corr(),vmin=0,vmax=1, annot=True)
```

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.set_xlabel('steps taken', fontsize = 30)
ax.set_ylabel('Calories burnt', fontsize=30)
ax.scatter(dailyActivity_merged['TotalSteps'], dailyActivity_merged['Calories'])
```

### Inferenes
##### 1> There are few outliers where steps taken is zero, while calories burned is greater than zero
##### 2> another outlier has very less calories burnt for maximum steps
##### 3> a general increase is seen for calories burnt, w.r.t steps taken

```python
#plt.subplots(figsize=(16,8))
plt.figure(figsize=(6,6))
plt.pie([sum(dailyActivity_merged['VeryActiveDistance']),sum(dailyActivity_merged['ModeratelyActiveDistance']),sum(dailyActivity_merged['LightActiveDistance']),sum(dailyActivity_merged['SedentaryActiveDistance'])],labels = ['Very Active Distance','Moderately Active Distance','Light Active Distance','Sedentary Active Distance'],wedgeprops = {"edgecolor": "white"})
plt.title('Distances travelled overall', fontsize=30)
plt.show()
```

### Inference
#### 1> it is observed that users only travel light distances, followed by very active distances and moderately active distances.

```python
plt.style.use("default")
plt.figure(figsize=(6,4))
plt.hist(dailyActivity_merged.Day, bins = 7, width = 0.55, color = "cyan", edgecolor = "brown")
plt.title('Frequency of app usage (vs) day of the week')
```

### Inference
#### 1> It is observed that the app is more frequently used on weekdays rather than the weekends
